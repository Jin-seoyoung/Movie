```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(all_Movie, random_state=0, test_size=0.2)
features = ['time', 'dis_rank_num', 'num_staff', 'dir_prev_bfnum', 'dir_prev_num', 'genre_num', 'screening_rat']
target = ['box_off_num']

X_train, y_train = train[features], np.log1p(train[target])
test_set, check_set = test[features], test[target]
```

```python
X_train = pd.get_dummies(columns = ['screening_rat'], data = X_train)
test_set = pd.get_dummies(columns = ['screening_rat'], data = test_set)
```

```python
X_train
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import KFold
from sklearn.ensemble import RandomForestRegressor
```

```python
kf = KFold(n_splits = 10, shuffle = True, random_state=42)
```

```python
rf = RandomForestRegressor(random_state = 518)
```

```python
rf_pred = np.zeros((test.shape[0]))
rf_pred
```

```python
rf_pred = np.zeros((test.shape[0]))

#kf.split을 통해 10개의 훈련 데이터 세트로 나누고 10번의 학습을 거침
#np.expm1은 로그 함수 numpy.log1p()로 변환 된 값을 원래 값으로 변환

for tr_idx, val_idx in kf.split(X_train, y_train) :
    tr_x, tr_y = X_train.iloc[tr_idx], y_train.iloc[tr_idx]
    
    rf.fit(tr_x, tr_y)
    
    sub_pred = np.expm1([0 if x < 0 else x for x in rf.predict(test_set)])

    rf_pred += (sub_pred/10) 
```

```python
from sklearn.metrics import mean_squared_error
def RMSE(check_set, rf_pred):
    return np.sqrt(mean_squared_error(check_set, rf_pred))
print('RMSE : ', RMSE(check_set, rf_pred)) 
```

```python
from sklearn.model_selection import GridSearchCV
 
param_grid = [
        {'n_estimators': [3, 10, 30], 'max_features': [2, 4, 6, 8]},
        {'bootstrap': [False], 'n_estimators': [3, 10], 'max_features': [2, 3, 4]},
    ]
forest_reg = RandomForestRegressor()
 
grid_search = GridSearchCV(forest_reg, param_grid, cv=5,
                           scoring='neg_mean_squared_error',
                           return_train_score=True)
 
grid_search.fit(X_train, y_train)
```
